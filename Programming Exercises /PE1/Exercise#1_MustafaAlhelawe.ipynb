```python
for row in fetal_dirty_df.index:
    if(fetal_dirty_df.loc[row].isna().sum() > 0): #checks each row for missing values
        print("Missing value in sample",row, ":" , fetal_dirty_df.loc[row].isna().sum())
```

### Problem 1.3

```python
fddf = fetal_dirty_df.copy()
for row in fddf.index:
    if (fddf.at[row,'baseline value'] < 0): #checks if value at specific row,column is negative
        fddf.at[row,'baseline value'] = None
fddf.loc[91:100]
```

I used the '.hist()' function to compare the histograms of each column in this dataset with that of the clean dataset, and the only inconsistency that I noticed was in the 'baseline value' column. The dirty dataset contained a few negative values, which is logically inconsistent as heart rate cannot be negative. The code above went through all the values in that column and replaced the negative values with 'NaN'.

### Problem 1.4

```python
fddfshift = sps.zscore(fddf.loc[:, fddf.columns!='fetal_health'], nan_policy = 'omit')#calculates zscore and negates categorical variables, and omits NaN values
fddfshift.loc[:, ['fetal_health']] = fddf[['fetal_health']] #adds fetal_health column to the new shifted dataset
fddfshift
```

### Problem 1.5

```python
fddf2 = fddfshift.copy()
for x in fddf2.columns:
    fddf2[x] = fddf2[x].fillna(fddf2[x].median()) #fills the NaN values with the median
    
fddf2.loc[91:100]
```

### Problem 1.6

```python
pd.get_dummies(data=fddf2, columns=['fetal_health']).to_csv('fetal_health_dataset_processed.csv', index = False)
```

```python
hdf = heart_df.copy()
for row in hdf.index:
    if (hdf.at[row,'serum_sodium'] < 120 and hdf.at[row, 'DEATH_EVENT'] == 0):
        hdf.at[row,'serum_sodium'] = None
hdf.loc[150:200]
```

## Problem 2.3

```python
hdfshift = sps.zscore(hdf.loc[:, ~hdf.columns.isin(['anaemia',
                                                    'diabetes',
                                                    'high_blood_pressure',
                                                    'sex',
                                                    'smoking',
                                                    'DEATH_EVENT'])], nan_policy = 'omit')
hdfshift.loc[:, ['anaemia', 'diabetes', 'high_blood_pressure', 'sex', 'smoking', 'DEATH_EVENT']] = hdf[['anaemia', 'diabetes', 'high_blood_pressure', 'sex', 'smoking', 'DEATH_EVENT']]
hdf2 = hdfshift.iloc[:, [0, 7, 1, 8, 2, 9, 3, 4, 5, 10, 11, 6, 12]]
hdf2.head(20)
```
